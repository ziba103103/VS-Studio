### Создал датафрейм, обозначил клонки в соответствии со значением 

```python
import pandas as pd 


treder = pd.read_csv('C:\\Users\\Admin\\Desktop\\Data_pandas\\EURUSDH2_cleaned.csv', header=None)
# Указание имен колонок
column_names = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Status']
#treder.columns = column_names
treder['Date'] = treder[0].apply(lambda X: X.split(' ')[0])
treder['Time'] = treder[0].apply(lambda X: X.split(' ')[1])
treder['Open'] = treder[0].apply(lambda X: X.split(' ')[2])
treder['High'] = treder[0].apply(lambda X: X.split(' ')[3])
treder['Low'] = treder[0].apply(lambda X: X.split(' ')[4])
treder['Close'] = treder[0].apply(lambda X: X.split(' ')[5])
treder['Volume'] = treder[0].apply(lambda X: X.split(' ')[6])
treder['Status'] = treder[0].apply(lambda X: X.split(' ')[7])
treder = treder.drop(columns=[0])

treder
```
